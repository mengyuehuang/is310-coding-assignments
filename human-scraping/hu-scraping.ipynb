```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import csv
```

```python
response = requests.get('https://humanist.kdl.kcl.ac.uk/')
soup = BeautifulSoup(response.text, features="html.parser")

links = soup.find_all('a')

# leave the speical case of Virginia/v01/ out
filtered_links = [link for link in links if str(link) != '<a href="/Archives/Virginia/v01/">Volume 1<br/> 5/87-5/88</a>']
filtered_links
```

```python
with open('humanist_volumes_original.csv', 'w') as file:
    writer = csv.writer(file)
    # Write the headers
    writer.writerow(['Link', 'Volume Text'])
    for link in filtered_links:
        if "Archives" in link.get('href'):
            volume_url = "https://humanist.kdl.kcl.ac.uk" + link.get('href')
            volume_response = requests.get(volume_url)
            volume_soup = BeautifulSoup(volume_response.text, "html.parser")
            writer.writerow([link.get_text(), volume_soup.get_text()[:1000]])

original = pd.read_csv('humanist_volumes_original.csv')
original
```

```python
original2 = original
original2['Volume Text'] = original2['Volume Text'].apply(lambda x: x.replace('\r', '').replace('\n', '').replace('\t', ''))
original2
```

```python
original2 = original2.drop([20, 21]) # dropping the rows we do not want
```
