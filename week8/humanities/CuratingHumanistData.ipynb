```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import csv
```

```python
response = requests.get('https://humanist.kdl.kcl.ac.uk/Archives/')
soup = BeautifulSoup(response.text, features="html.parser")

# the first link is the link to the parent directory, we do not need it
mother_links = soup.find_all('a')[1:]
mother_links
```

#### Now, the normal scraping with each link pile

1st one

```python
most_outter_link1 = "https://humanist.kdl.kcl.ac.uk/Archives/" + mother_links[0]['href']
most_outter_link1
```

```python
volume1_response = requests.get(most_outter_link1)
volume1_soup = BeautifulSoup(volume1_response.text, features="html.parser")

pile1_links = volume1_soup.find_all('a')
pile1_links
```

```python
with open('csv1.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    # Write the headers
    writer.writerow(['Link', 'Volume Text'])
    for link in pile1_links:
        # change the checker to only scrape the volumes
        if "humanist" in link.get('href'):
            volume_url = most_outter_link1 + link.get('href')
            volume_response = requests.get(volume_url)
            volume_soup = BeautifulSoup(volume_response.text, "html.parser")
            writer.writerow([link.get_text(), volume_soup.get_text()[:10000]])

csv1 = pd.read_csv('csv1.csv')
csv1
```

and it has an extra layer (compared with previous 2) as nested webs

```python
with open('csv3.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    # Write the headers
    writer.writerow(['Link', 'Volume Text'])


    # checked above
    nested_url = most_outter_link3 + pile3_links[0].get('href')

    nested_response = requests.get(nested_url)
    nested_soup = BeautifulSoup(nested_response.text, features="html.parser")

    nested_links = nested_soup.find_all('a')


    for link in nested_links:
        if "volume" in link.get('href'):
            volume_url = nested_url + link.get('href')
            volume_response = requests.get(volume_url)
            volume_soup = BeautifulSoup(volume_response.text, "html.parser")
            writer.writerow([link.get_text(), volume_soup.get_text()[:10000]])

csv3 = pd.read_csv('csv3.csv')
csv3
```
