```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import csv
```

```python
sources = pd.read_csv('cleaned_pudding_data.csv')
sources
```

```python
with open('pudding.csv', 'w') as file:
    writer = csv.writer(file)
    writer.writerow(['Volume Text'])  # headers

    # loop through the sources
    for source in sources['link']:
        response = requests.get(source)
        soup = BeautifulSoup(response.text, features="html.parser")
        # write the text to the file
        writer.writerow([soup.get_text()[:1000]])

pudding = pd.read_csv('pudding.csv')
pudding
```

```python
pudding['Volume Text'] = pudding['Volume Text'].apply(lambda x: x.replace('\r', '').replace('\n', '').replace('\t', ''))
pudding
```

```python
pudding1 = pudding
pudding1['title'] = sources['title']
pudding1 = pudding1[~pudding1['Volume Text'].str.contains('Forbidden')]
pudding1
```

```python
pudding1.to_csv('pudding_movie_dialogue.csv', index=False)
```
